```python
import pandas as pd
#importing pandas library
```

```python
dataset=pd.read_csv("Salary_Data.csv")
#reading the data file using read_csv function in pandaas library
```

```python
dataset
```

```python
independent=dataset[["YearsExperience"]]
#assingning the input column to variable named independent
```

```python
independent
```

```python
dependent=dataset[["Salary"]]
#assigning the output column to a variable named dependent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
#dataset splitting in to test and train datas using train test function
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
#model creation by giving the training input and output values
```

```python
weight=regressor.coef_
weight
#assigning the weight value stored in the Linearregression class to a variable named 'weight' by calling the corresponding instance variable through a object.
```

```python
bais=regressor.intercept_
bais
#assigning the bias value stored in the Linearregression class to a variable named 'bias' by calling the corresponding instance variable through a object.
```

```python
Y_pred=regressor.predict(X_test)
#predicting the output with the created model by giving the test input and assigning the predicted output to a variable named 'Y_pred'
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
#evaluating the test and predicted output to obtain the r2 score
```

```python
r_score
```

```python
import pickle
filename="finalised_model_linear.sav"
#creation of new file to store the model and assigning it to a variable 'filename'.
```

```python
pickle.dump(regressor,open(filename,'wb'))
#dumping the final model in the regressor object to the created file.
```

```python
loaded_model=pickle.load(open("finalised_model_linear.sav",'rb'))
#loading the model from the stored file in to a variable....(loading process to be created in new notebook)
result=loaded_model.predict([[13]])
#calling the predict function in the model and feeding them with the new input values and storing the predicted output in a variable named 'result'
```
